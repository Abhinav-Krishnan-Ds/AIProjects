```python
model_dir = "/home/abhinav/Documents/Work/2 Hobby_projects/Models"

path_to_saved_arch = os.path.join(model_dir, "Mynet_beetles_7")
```

```python
model = load_model(path_to_saved_arch)
```

```python
# importing the necessary libraries
import cv2
import numpy as np

# Creating a VideoCapture object to read the video
cap = cv2.VideoCapture(0)


# Loop until the end of the video
while (cap.isOpened()):

    # Capture frame-by-frame
    ret, frame = cap.read()
    frame = cv2.resize(frame, (456, 356), fx = 0, fy = 0,
                         interpolation = cv2.INTER_CUBIC)

    # Display the resulting frame
    
    frame = cv2.flip(frame, 1)
    nframe = tf.image.resize(frame, [140, 140])
    
    nframe = tf.expand_dims(nframe, 0)
    # conversion of BGR to grayscale is necessary to apply this operation
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # adaptive thresholding to use different threshold 

    Thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                           cv2.THRESH_BINARY_INV, 11, 2)
    prediction = model.predict(nframe)
    print("Shape of prediction")
    print(prediction)
    prediction_values = np.argmax(prediction, axis=-1)
    print("Shape of prediction_Values: ", prediction_values.shape)
    print("prediction values: ", prediction_values)
    # c = []
    # for i in prediction_values:
    #     c.append(class_names[i])
    # print(c)

    # prediction_act_values = c[0]
    # print(prediction_act_values)
    if prediction_values[0] == 2:
        cv2.putText(frame,"Beetle Detected",(50,50),cv2.FONT_HERSHEY_COMPLEX_SMALL,2,(100,0,255))
    else:
        cv2.putText(frame, str(""),(50,50),cv2.FONT_HERSHEY_COMPLEX_SMALL,2,(100,0,255))

    cv2.imshow('Frame', frame)
    # cv2.imshow('Thresh', Thresh)
    # cv2.imshow(prediction_act_values, Thresh)
    # define q as the exit button
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

# release the video capture object
cap.release()
# Closes all the windows currently opened.
cv2.destroyAllWindows()
```
